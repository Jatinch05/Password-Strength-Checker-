```python
import matplotlib.pyplot as plt
import numpy as np

# Given data
data = [92, 62, 22, 75, 55, 63, 85, 69, 72, 84]

# Plotting the histogram
plt.figure(figsize=(10, 6))

# Histogram plot
plt.subplot(1, 2, 1)
plt.hist(data, bins=6, edgecolor='black', alpha=0.7)
plt.title('Histogram of Data')
plt.xlabel('Values')
plt.ylabel('Frequency')

# Box plot
plt.subplot(1, 2, 2)
plt.boxplot(data, vert=False)
plt.title('Box Plot of Data')
plt.xlabel('Values')

# Show the plots
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Sample noisy data
x = np.linspace(0, 10, 100)
y = np.sin(x) + np.random.normal(scale=0.3, size=x.shape)

# Polynomial regression for smoothing
poly = PolynomialFeatures(degree=3)
X_poly = poly.fit_transform(x.reshape(-1, 1))
model = LinearRegression()
model.fit(X_poly, y)

# Predicted (smoothed) values
y_smooth = model.predict(X_poly)

# Plot the original noisy data and smoothed data
plt.scatter(x, y, label='Noisy Data', alpha=0.5)
plt.plot(x, y_smooth, color='red', label='Smoothed Data')
plt.legend()
plt.title('Data Smoothing Using Regression')
plt.show()
```
